```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/dancbatista/projeto_final_awari/main/datahackers-survey-2019-anonymous-responses.csv")
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df.groupby(["('P8', 'degreee_level')", "('P2', 'gender')"]).count()["('P0', 'id')"]
```

```python
df.groupby(["('P10', 'job_situation')", "('P2', 'gender')"]).count()["('P0', 'id')"].plot(kind="barh")
```

```python
df["('P16', 'salary_range')"].value_counts()
```

```python
salary_order = ['Menos de R$ 1.000/mês','de R$ 1.001/mês a R$ 2.000/mês',
                'de R$ 2.001/mês a R$ 3000/mês','de R$ 3.001/mês a R$ 4.000/mês',
                'de R$ 4.001/mês a R$ 6.000/mês','de R$ 6.001/mês a R$ 8.000/mês',
                'de R$ 8.001/mês a R$ 12.000/mês','de R$ 12.001/mês a R$ 16.000/mês',
                'de R$ 16.001/mês a R$ 20.000/mês','de R$ 20.001/mês a R$ 25.000/mês',
                'Acima de R$ 25.001/mês',np.nan]
```

```python
plt.figure(figsize=(7,5))
sns.countplot(y="('P16', 'salary_range')",hue="('P2', 'gender')",order=salary_order, data=df)
```

```python
plt.figure(figsize=(7,11))
sns.countplot(y="('P16', 'salary_range')",hue="('P22', 'most_used_proggraming_languages')",order=salary_order, data=df)
```
