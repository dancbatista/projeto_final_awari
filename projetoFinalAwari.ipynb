```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/dancbatista/projeto_final_awari/main/datahackers-survey-2019-anonymous-responses.csv")
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df.groupby(["('P8', 'degreee_level')", "('P2', 'gender')"]).count()["('P0', 'id')"]
```

```python
df.groupby(["('P10', 'job_situation')", "('P2', 'gender')"]).count()["('P0', 'id')"].plot(kind="barh")
```

```python
df["('P16', 'salary_range')"].value_counts()
```

```python
salary_order = ['Menos de R$ 1.000/mês','de R$ 1.001/mês a R$ 2.000/mês',
                'de R$ 2.001/mês a R$ 3000/mês','de R$ 3.001/mês a R$ 4.000/mês',
                'de R$ 4.001/mês a R$ 6.000/mês','de R$ 6.001/mês a R$ 8.000/mês',
                'de R$ 8.001/mês a R$ 12.000/mês','de R$ 12.001/mês a R$ 16.000/mês',
                'de R$ 16.001/mês a R$ 20.000/mês','de R$ 20.001/mês a R$ 25.000/mês',
                'Acima de R$ 25.001/mês',np.nan]
```

```python
plt.figure(figsize=(7,11))
sns.countplot(y="('P16', 'salary_range')",hue="('P22', 'most_used_proggraming_languages')",order=salary_order, data=df)
```

```python
plt.figure(figsize=(7,5))
sns.countplot(y="('P16', 'salary_range')",hue="('P2', 'gender')",order=salary_order, data=df)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
fig.suptitle('Banco de dados mais Usados')

# MYSQL
sns.countplot(ax=axes[0], x="('P26', 'mysql')", data=df)

# Oracle
sns.countplot(ax=axes[1], x="('P26', 'oracle')", data=df)

# SQL_SERVER
sns.countplot(ax=axes[2], x="('P26', 'sql_server')", data=df)
```

```python
df["('P24', 'imagens')"].value_counts().plot(kind="bar")
```

```python
df["('P17', 'time_experience_data_science')"].unique()

time_experience_order = ['Não tenho experiência na área de dados', 'Menos de 1 ano', 'de 1 a 2 anos', ' de 2 a 3 anos', 'de 4 a 5 anos', 'de 6 a 10 anos', 'Mais de 10 anos', np.nan]
```

```python
plt.figure(figsize=(7,11))
sns.countplot(y="('P10', 'job_situation')",hue="('P17', 'time_experience_data_science')", hue_order=time_experience_order, data=df)
```

```python
plt.figure(figsize=(7,11))
sns.countplot(y="('P16', 'salary_range')",hue="('P17', 'time_experience_data_science')",order=salary_order, hue_order=time_experience_order, data=df)
```

```python
df["('P18', 'time_experience_before')"].unique()
```

```python
experience_before_order = ['Não tive experiência na área de TI/Engenharia de Software antes de começar a trabalhar na área de dados', 
                           'Menos de 1 ano', 'de 1 a 2 anos', 'de 2 a 3 anos', 'de 4 a 5 anos', 'de 6 a 10 anos', 'Mais de 10 anos', np.nan]
```

```python
data_scientist = df[df["('P19', 'is_data_science_professional')"] == 1]
```

```python
no_data_scientist = df[df["('P19', 'is_data_science_professional')"] == 0]
```

```python
plt.figure(figsize=(7,11))
sns.countplot(y="('P16', 'salary_range')",hue="('P18', 'time_experience_before')",order=salary_order, hue_order=experience_before_order, data=data_scientist)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharey=True)
fig.suptitle('Salários')

# No_Data_Scientist
sns.countplot(ax=axes[0], y="('P16', 'salary_range')", data=no_data_scientist)

# Data_Scientist
sns.countplot(ax=axes[1], y="('P16', 'salary_range')", data=data_scientist)

```

```python
plt.figure(figsize=(7,11))
sns.countplot(y="('P16', 'salary_range')",hue="('P18', 'time_experience_before')",order=salary_order, hue_order=time_experience_order, data=no_data_scientist)
```
